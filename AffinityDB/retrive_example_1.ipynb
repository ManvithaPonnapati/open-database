## Data retrive example

create a retrive data class

load different kinds of data from different table by call the api from the class created.

export the selected result as a pandas.DataFrame

```python
from retrive import retrive_data
```

### create a class to retrive data

```python
ra = retrive_data()
```

### Set receptor
Load the information for splited receptors from table 2

select the receptor with X-Ray resolution smaller than 3

```python
ra.receptor(2,(None, 3))
```

### Set ligand
Get the information for reordered ligand from table 3

```python
ra.crystal(4)
```

### Exclude ligands
Get the exclued ligand info from table 10

```python
ra.exclusion(10)
```

### Select ligan in the box

Get the ligand size info form table:3

select the ligand which the largest atom pair distance along axis x,y or z no more than 20 A

```python
ra.ligand_size(3,[None, 20])
```

### Select ligand by affinity
Get the information for binding affintiy from table 9

select the ligand with log binding affinity value in [-30, -20]

### rest 
- tuple: (a,b) , restriction a < value < b 
- list: [a,b] , restriction a <= value <= b
- None: no restriction


```python
ra.log_affinity(9, rest=[-30,-20])
```

### export 
export the selected ligand as pandas dataframe

```python
db_table = ra.export_table()
```

```python
db_table
```

Export the selected data to the folder config.export_dir/test_tfr_one
